```python
plt.scatter(x,y)
plt.xlabel("Advertising cost")
plt.ylabel("Sales")
plt.show()
```

```python
slope = 0.7
intercept = -0.1
def myfunc(x):
    return slope*x+intercept
mymodel = list(map(myfunc,x))
```
